## Flight Price Prediction ( EDA + Feature Engineering)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_excel('Data_Train.xlsx')
df_test = pd.read_excel('Test_set.xlsx')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
df.head()
```

```python
#Feature Engineering process
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.head()
```

```python
df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']= df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Route'].mode()[0]
```

```python
df['Route'] = df['Route'].fillna(df['Route'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna(df['Total_Stops'].mode()[0])
```

```python
df.isnull().sum()  
```

```python
df.head()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df.head()
```

```python
df.drop(['Arrival_Time'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Departure_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Departure_Minute'] = df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.head()
```

```python
df.drop(['Dep_Time'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4})
```

```python
df.head()
```

```python
df.drop(['Route'], axis=1, inplace=True)
```

```python
df.head()
```

```python
#FEATURE SCALING
```

```python
from sklearn.preprocessing import StandardScaler
cols_to_scale = ['Price', 'Total_Stops', 'Arrival_Hour', 'Arrival_Minute', 'Departure_Hour', 'Departure_Minute']
scaler = StandardScaler()

# Fit and transform selected columns
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```
