```python
df_country = pd.read_excel('Country-Code.xlsx') #read country code data
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left') #merge dataframes on 'Country Code'
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_counts = final_df.Country.value_counts().index #count of unique values in 'Country Code'
country_counts
```

```python
country_values = final_df.Country.value_counts().values #count of unique values in 'Country Code'
country_values
```

```python
#Pie chart of top 3 countries by count
plt.pie(country_values[:3], labels=country_counts[:3], autopct='%1.2f%%', shadow=True)
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'}) #.reset_index() converts the groupby object to a DataFrame
ratings
```

```python
#Conclusions
# 1. Whenever the rating is from 4.5 to 5, the rating color is 'Dark Green' and the rating text is 'Excellent'.
# 2. Whenever the rating is from 4 to 4.5, the rating color is 'Green' and the rating text is 'Very Good'.
# 3. Whenever the rating is from 3.5 to 4, the rating color is 'Yellow' and the rating text is 'Good'.
# 4. Whenever the rating is from 2.5 to 3.5, the rating color is 'Orange' and the rating text is 'Average'.
# 5. Whenever the rating is less than 2.5, the rating color is 'Red' and the rating text is 'Poor'.
```

```python
ratings.head(3)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', palette=['Blue','Red','Orange','Yellow','Green','Cyan']) #bar plot of ratings
```

```python
#Observations
# 1. There are no restaurants with an aggregate rating of 0.
# 2. The maximum number of restaurants have an aggregate rating of 4.0.
```

```python
sns.countplot(x='Rating color', data=ratings, palette=['Blue','Red','Orange','Yellow','Green','Cyan']) #count plot of rating colors
```
